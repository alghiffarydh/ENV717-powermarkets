## Question 1

```python
#import the excel file from the website
egrid_plnt20=pd.read_excel("https://www.epa.gov/system/files/documents/2022-01/egrid2020_data.xlsx", 
                       sheet_name="PLNT20", keep_default_na=False, header=1)
egrid_plnt20.head(10) 
```

```python
#Name of columns
# Plant name: PNAME
# Plant state abbreviation: PSTATABB
# Utility name: UTLSRVNM
# Plant-level sector: SECTOR
# Balancing Authority Name: BANAME
# Plant associated ISO/RTO Territory: ISORTO
# Plant capacity factor: CAPFAC
# Plant nameplate capacity: NAMEPCAP
# Plant annual net generation (MWh): PLNGENAN

#Change the data to numeric
egrid_plnt20.CAPFAC=pd.to_numeric(egrid_plnt20.CAPFAC)
egrid_plnt20.NAMEPCAP=pd.to_numeric(egrid_plnt20.NAMEPCAP)
egrid_plnt20.PLNGENAN=pd.to_numeric(egrid_plnt20.PLNGENAN)

#Create a new data frame for question 1
egrid_plnt20_q1 = egrid_plnt20.loc[:,['PNAME', 'PSTATABB', 'UTLSRVNM','SECTOR','BANAME', 
                              'ISORTO', 'CAPFAC','NAMEPCAP', 'PLNGENAN']].copy()

egrid_plnt20_q1.head(n=10)
```

```python
#Remove the plants with zero or blank capacity factor from the data frame
egrid_plnt20_q1.dropna(axis='rows', subset=['CAPFAC'],inplace=True)
egrid_plnt20_q1 = egrid_plnt20_q1[(egrid_plnt20_q1.CAPFAC > 0)].copy()
```

```python
#Label the plants that belong to an ISO/RTO as Yes and No for vice versa
egrid_plnt20_q1['is_iso_rto'] = 'Yes'
egrid_plnt20_q1.loc[egrid_plnt20_q1.ISORTO == '', 'is_iso_rto'] = 'No'
#Create a pivot table by state, with nameplate capacity and total generation 
#Values aggregated by plants that belong to ISO/RTO or not
plants_iso = pd.pivot_table(egrid_plnt20_q1, index="PSTATABB", columns='is_iso_rto', 
                        values=['NAMEPCAP', 'PLNGENAN'], fill_value=0, aggfunc=['sum'])
#Drop index level in columns 
plants_iso.columns = plants_iso.columns.droplevel(level =0)
plants_iso.head(n=10)
```

```python
#Determine the states that are considered as ISO/RTO states based on the dominant percentage of plants under/not under ISO/RTO
plants_iso['percent_iso'] = 100*plants_iso[('NAMEPCAP','Yes')]  / (plants_iso[('NAMEPCAP','Yes')] +plants_iso[('NAMEPCAP','No')] )
#Add and initialize two columns to the pivot table
#The columns "Number of States" will be useful later to count how many have restructured 
plants_iso['Number of States'] = 1
#The column "restructured" is to indicate for each state if it is restructured or not. Initalize in no and change later based on % of ISO/RTO
plants_iso['restructured'] = 'N'
plants_iso.loc[plants_iso.percent_iso >= 51, 'restructured'] = 'Y'

#Include capacity and generation by adding both in and outside ISO/RTO
plants_iso['Capacity in 2020 (MW)'] = plants_iso[('NAMEPCAP','Yes')] +plants_iso[('NAMEPCAP','No')]
plants_iso['Generation in 2020 (MWh)'] = plants_iso[('PLNGENAN','Yes')] +plants_iso[('PLNGENAN','No')]

#Group data by the states that are mostly reformed and not reformed
plants_iso_sum = plants_iso.groupby(['restructured']).sum()
plants_iso_sum.columns = plants_iso_sum.columns.droplevel(level =1)
plants_iso_sum.rename(index={'Y': 'RTO/ISO', 'N':'Non-RTO/ISO'}, inplace=True)
plants_iso_sum.index.name = 'Region'
plants_iso_sum[['Number of States',	'Capacity in 2020 (MW)','Generation in 2020 (MWh)']]
```
